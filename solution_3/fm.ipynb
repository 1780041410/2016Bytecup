```python
import graphlab as gl
```

```python
train_all=gl.SFrame.read_csv("train_1115_huidalv_improve.csv")
```

```python
train_all['huidalv_cost']=train_all.apply(lambda x: x['huidalv']/1.3 if x['huidalv']>=0.4 and x['zj_answers']*90<10 else x['huidalv'] )
train_all['label_huidalv_cost']=train_all.apply(lambda x: x['label_huidalv']/1.3 if x['label_huidalv']>=0.4 
                                                and x['zj_q_label_answers']*90<10 
                                                else x['label_huidalv'] )
```

```python
validate_all=gl.SFrame.read_csv("validate_1115_for_predict.csv")
```

```python
validate_all['huidalv_cost']=validate_all.apply(lambda x: x['huidalv']/1.3 if x['huidalv']>=0.4 and x['zj_answers']*90<10 else x['huidalv'] )
validate_all['label_huidalv_cost']=validate_all.apply(lambda x: x['label_huidalv']/1.3 if x['label_huidalv']>=0.4
                                                      and x['zj_q_label_answers']*90<10 
                                                      else x['label_huidalv'] )
```

```python
validate_all
```

```python
test_all=gl.SFrame.read_csv("test_1115_for_predict.csv")
```

```python
test_all['huidalv_cost']=test_all.apply(lambda x: x['huidalv']/1.3 if x['huidalv']>=0.4 and x['zj_answers']*90<10 else x['huidalv'] )
test_all['label_huidalv_cost']=test_all.apply(lambda x: x['label_huidalv']/1.3 if x['label_huidalv']>=0.4
                                                      and x['zj_q_label_answers']*90<10 
                                                      else x['label_huidalv'] )
```

```python
user_data =train_all[['uid','huidalv_cost']] 
```

```python
item_data=train_all[['qid','q_label','q_sp']] 
```

```python
train_gl=train_all[['uid','qid','label','label_huidalv_cost']]
validate_gl=validate_all[['uid','qid','label_huidalv_cost']]
test_gl=test_all[['uid','qid','label_huidalv_cost']]
```

```python
m = gl.recommender.factorization_recommender.create(train_gl,user_id='uid',item_id='qid',
                                                            user_data=user_data,item_data=item_data,
                                                            target='label',
                                                            #num_sampled_negative_examples =50,
                                                            max_iterations=3000,solver='auto',
                                                            binary_target=True,
                                                            linear_regularization =0.006,regularization =0.006,
                                                            side_data_factorization =True,
                                                            #ranking_regularization =0.004,
                                                            num_factors=1024,                      
                                                            #unobserved_rating_value =0,
                                                            random_seed=500 )
m.save('m1118_huidalv_improve_cost_2.model')
```

```python
#m=gl.load_model('m1115_huidalv_improve_tiaocan.model')
```

```python
validate_user_data =validate_all[['uid','huidalv_cost']] 
validate_item_data =validate_all[['qid','q_label','q_sp']] 
```

```python
test_user_data =test_all[['uid','huidalv_cost']] 
test_item_data =test_all[['qid','q_label','q_sp']] 
```

```python
test_recs = m.predict(test_gl,new_user_data=test_user_data,new_item_data=test_item_data)
```

```python
recs = m.predict(validate_gl,new_user_data=validate_user_data,new_item_data=validate_item_data)
```

```python
recs
```

```python
validate_gl.add_column(recs,name='label')
```

```python
test_gl.add_column(test_recs,name='label')
```

```python
output=validate_gl[['qid','uid','label']]
output.save('output_validate1118_huidalv_cost_2.csv') 
```
