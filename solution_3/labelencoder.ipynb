```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelBinarizer
```

```python
user=pd.read_table('data/user_info.txt',sep='\t',names=['uid','ulabel','uword','ualpha'],index_col='uid')
user['uid']=user.index
user.head(2)
```

```python
uid_ulabel=[]
a=user.apply(lambda x: [uid_ulabel.append([x['uid'],l]) for l in x['ulabel'].split('/')], axis=1)
df_uid_ulabel=pd.DataFrame(uid_ulabel,columns=['uid','ulabel'])

lb=LabelBinarizer()
lb.fit(df_uid_ulabel['ulabel'])
lb_ulabel=lb.transform(df_uid_ulabel['ulabel'])

_columns=['l'+str(i) for i in range(lb_ulabel.shape[1])]
df_lb_ulabel=pd.concat([df_uid_ulabel[['uid']],pd.DataFrame(lb_ulabel,columns=_columns)],axis=1)
df_lb_ulabel.head(5)
```

```python
keys = df_lb_ulabel.groupby('uid').groups
u_labels=[]
for key in keys:
    pred=df_lb_ulabel[df_lb_ulabel['uid']==key][_columns]
    r=pred.sum(axis=0)
    rr=[]
    rr.append(key)
    temp=[rr.append(j) for j in r]
    u_labels.append(rr)
    
ulabel_finall=pd.DataFrame(u_labels)
_columns=['uid']
_columns.extend(['l'+str(i) for i in range(ulabel_finall.shape[1]-1)])
ulabel_finall.columns=_columns
ulabel_finall.head()
```

```python
# 把DATAFRAME转换成FM的形式
def get_fm_data(data):
#     data.columns=['x'+str(i) for i in range(train_data.shape[1])]
    x_data=[]
    for i in range(data.shape[0]):
        x_d={}
        for j in range(data.shape[1]):
            x_d[str(j)]=data.ix[i,j]
        x_data.append(x_d)
    return x_data
```
